```python
banzuke_df = pd.read_excel('C:/Users/anush/Desktop/Sumo Wrestling/Banzuke_Final.xlsx')
```

```python
my_dict = dict(zip(banzuke_df.Rikishi, banzuke_df.Id))
```

```python
new_df = pd.read_excel("C:/Users/anush/Desktop/Banzuke_Shikona_Filter.xlsx")
```

```python
for key, value in my_dict.items():
    for state in new_df['New_Name']:
        if key == state:
            new_df.loc[(new_df['New_Name'] == key), "New_Name_id"] = value
            break
```

```python
new_df.to_excel("C:/Users/anush/Desktop/Banzuke_Shikona_Filter.xlsx")
```
